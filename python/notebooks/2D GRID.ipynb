
Complexity analysis variables:

````python
class WordFinder:
    def __init__(self,grid, word):
        self.grid = grid
        self.max_rows = len(grid)
        self.max_cols = len(grid[0])
        self.word_chars = list(word)
        
    def word_finder(self, row = 0, col = 0, letters = []):
        print(self.grid[row][col]+'----'+self.word_chars[len(letters)] +'-------('+str(row)+','+str(col)+')----'+ str(self.grid[row][col] == self.word_chars[len(letters)]))
        if self.grid[row][col] == self.word_chars[len(letters)]:
            letters = letters + [(row,col)]
        else:
            letters = []
        
            
        if len(letters) == len(self.word_chars):
            return letters
        else:
            new_letters = []
            #Move towards the column that contains the required character 
            if col+1 < self.max_cols and self.word_chars[len(letters)] == self.grid[row][col+1]:
                new_letters = self.word_finder(row, col+1, letters)
            #Move towards the row that contains the required character
            if len(new_letters) == 0 and row+1 < self.max_rows and self.word_chars[len(letters)] == self.grid[row+1][col]:
                new_letters = self.word_finder(row+1, col, letters)
            #Increment the col and search for the letters in the next column until the last column
            if len(new_letters) == 0 and col+1 < self.max_cols:
                new_letters = self.word_finder(row,col+1, letters)
            #Increment the row up until the we reach the last column then break
            if len(new_letters) == 0 and row+1 <self.max_rows and col+1 < self.max_cols:
                new_letters = self.word_finder(row+1, col, letters)
            
            return new_letters
````
